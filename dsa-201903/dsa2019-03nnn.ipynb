```python
# Alterar o tipo da coluna StateHoliday de int para booleano
train_file.StateHoliday = train_file.StateHoliday != '0'
test_file.StateHoliday  = test_file.StateHoliday  != '0'
```

```python
# Utilizar a automatização da biblioteca fast.ai para acrescentar colunas time-series
add_datepart(train_file, "Date", drop=False)
add_datepart(test_file,  "Date", drop=False)
```

```python
# join_df é uma função para unir tabelas. Por padrão, faz-se um  left outer join
# do argumento à direita usando os campos fornecidos para cada tabela.
# O join no Pandas é feito usando merge
# O argumento de sufixos descreve a convenção de nomenclatura para campos duplicados. 
# Foram mantidos os nomes dos campos duplicados à esquerda e acrescentado sufixo "_y"
# aos que estão à direita.

def join_df(left, right, left_on, right_on=None, suffix='_y'):
    if right_on is None: right_on = left_on
    return left.merge(right, how='left', left_on=left_on, right_on=right_on, 
                      suffixes=("", suffix))
```

```python
# anexar os dados das lojas ao de teste e de treino
# a coluna store é utilizada como key para a junção
treino = join_df(train_file, store_file, "Store")
teste  = join_df(test_file,  store_file, "Store")
len(treino[treino.StoreType.isnull()]), len(teste[teste.StoreType.isnull()])
```

```python
# Valores missing serão tratados com um sinal arbitrário que não está nos dados
for df in (treino, teste):
    df['CompetitionOpenSinceYear']  = df.CompetitionOpenSinceYear.fillna(1900).astype(np.int32)
    df['CompetitionOpenSinceMonth'] = df.CompetitionOpenSinceMonth.fillna(1).astype(np.int32)
    df['Promo2SinceYear']           = df.Promo2SinceYear.fillna(1900).astype(np.int32)
    df['Promo2SinceWeek']           = df.Promo2SinceWeek.fillna(1).astype(np.int32)
```

```python
# Extrair características dos campos "CompetitionOpenSince" e "CompetitionDaysOpen"
for df in (treino, teste):
    df["CompetitionOpenSince"] = pd.to_datetime(dict(year=df.CompetitionOpenSinceYear, 
                                                     month=df.CompetitionOpenSinceMonth, day=15))
    df["CompetitionDaysOpen"] = df.Date.subtract(df.CompetitionOpenSince).dt.days
```

```python
# Dar consistência aos dados eliminando valores negativos e
# tratando datas com ano inferior a 1990
for df in (treino, teste):
    df.loc[df.CompetitionDaysOpen<0, "CompetitionDaysOpen"] = 0
    df.loc[df.CompetitionOpenSinceYear<1990, "CompetitionDaysOpen"] = 0
```

```python
# Criar campo "CompetitionMonthsOpen", limitando o máximo de 2 anos para inibir o
# número de categorias únicas
for df in (treino, teste):
    df["CompetitionMonthsOpen"] = df["CompetitionDaysOpen"]//30
    df.loc[df.CompetitionMonthsOpen>24, "CompetitionMonthsOpen"] = 24

# Visualizar o resultado
treino.CompetitionMonthsOpen.unique()
```

```python
for df in (treino, teste):
    df["Promo2Since"] = pd.to_datetime(df.apply(lambda x: Week(
        x.Promo2SinceYear, x.Promo2SinceWeek).monday(), axis=1).astype(pd.datetime))
    df["Promo2Days"] = df.Date.subtract(df["Promo2Since"]).dt.days
```

```python
for df in (treino, teste):
    df.loc[df.Promo2Days<0, "Promo2Days"] = 0
    df.loc[df.Promo2SinceYear<1990, "Promo2Days"] = 0
    df["Promo2Weeks"] = df["Promo2Days"]//7
    df.loc[df.Promo2Weeks<0, "Promo2Weeks"] = 0
    df.loc[df.Promo2Weeks>25, "Promo2Weeks"] = 25
    df.Promo2Weeks.unique()
```

```python
# Criar função get_elapsed para contagem cumulativa em um dataframe ordenado.
# A partir de um campo, a função irá iniciar o tempo de rastreamento desde a
# última ocorrência daquele campo. Quando o campo for visto novamente, o contador
# é definido como zero.
def get_elapsed(fld, pre):
    day1 = np.timedelta64(1, 'D')
    last_date = np.datetime64()
    last_store = 0
    res = []

    for s,v,d in zip(df.Store.values,df[fld].values, df.Date.values):
        if s != last_store:
            last_date = np.datetime64()
            last_store = s
        if v: last_date = d
        res.append(((d-last_date).astype('timedelta64[D]') / day1))
    df[pre+fld] = res
```

```python
# A função será aplicada às seguintes colunas:
colunas = ["Date", "Store", "Promo", "StateHoliday", "SchoolHoliday"]
```

```python
#df = train[columns]
df = train_file[colunas].append(test_file[colunas])
```

```python
fld = 'SchoolHoliday'
df = df.sort_values(['Store', 'Date'])
get_elapsed(fld, 'After')
df = df.sort_values(['Store', 'Date'], ascending=[True, False])
get_elapsed(fld, 'Before')
```

```python
fld = 'StateHoliday'
df = df.sort_values(['Store', 'Date'])
get_elapsed(fld, 'After')
df = df.sort_values(['Store', 'Date'], ascending=[True, False])
get_elapsed(fld, 'Before')
```

```python
fld = 'Promo'
df = df.sort_values(['Store', 'Date'])
get_elapsed(fld, 'After')
df = df.sort_values(['Store', 'Date'], ascending=[True, False])
get_elapsed(fld, 'Before')
```

```python
# Transforma date em index
df = df.set_index("Date")
```

```python
colunas = ['SchoolHoliday', 'StateHoliday', 'Promo']
```

```python
# Trata nulidade
for o in ['Before', 'After']:
    for p in colunas:
        a = o+p
        df[a] = df[a].fillna(0).astype(int)
```

```python
bwd = df[['Store']+colunas].sort_index().groupby("Store").rolling(7, min_periods=1).sum()
```

```python
fwd = df[['Store']+colunas].sort_index(ascending=False
                                      ).groupby("Store").rolling(7, min_periods=1).sum()
```

```python
# Em seguida, eliminar os índices Store agrupados na função window
bwd.drop('Store',1,inplace=True)
bwd.reset_index(inplace=True)
```

```python
fwd.drop('Store',1,inplace=True)
fwd.reset_index(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
# Juntar os dataframes
df = df.merge(bwd, 'left', ['Date', 'Store'], suffixes=['', '_bw'])
df = df.merge(fwd, 'left', ['Date', 'Store'], suffixes=['', '_fw'])
```
